```python
import pandas as pd
import numpy as np
df=pd.read_csv('day.csv')
# seperating the numeric data and categorical data
num_col= ['temp', 'hum', 'windspeed']
cat_var = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday' , 'weathersit' ]
#We apply Boxplot using boxplot() function on the numeric variables as shown below
df.boxplot(num_col)


```

```python
for x in ['windspeed']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

```python
for x in ["windspeed"]:
  q75,q25=np.percentile(df.loc[:,x],[75,25])
  intr_qr= q75-q25
  max= q75+(1.5*intr_qr)
  min= q25-(1.5*intr_qr)
  df.loc[df[x]<min,x]=np.nan
  df.loc[df[x]>max,x]=np.nan
df.isnull().sum()
df=df.dropna()
df.isnull().sum()
```
